```python
import os
import re
import pandas as pd
```

This notebook process the output of the following command

<pre>
python3 -m models.create_detection_model -s  2>&1 > create.log
</pre>

```python
with open("create.log", "r") as f:
    lines = f.readlines()
print("Lines", len(lines))
```

```python
results = []
model = None
num_params = None
macs = None

for line in lines:
    if "Model name:" in line:
        model = line.split("Model name:")[1].strip()

    if model is not None:
        if "Trainable params:" in line:
            num_params = eval(line.split("Trainable params:")[1].replace(",", ""))
        elif "Total mult-adds (Units.GIGABYTES):" in line:
            macs = line.split("Total mult-adds (Units.GIGABYTES):")[1].strip()

            idx = model.index("_")
            backbone = model[idx+1:].replace("_", " ").title()
            model = model[:idx].lower()
            model = {"fasterrcnn": "Faster R-CNN", "ssd": "SSD", "fcos": "FCOS", }.get(model, model.title())
            results.append({"Model": model, "Backbone": backbone, "Num. trainable params": num_params, "MACS": macs})
            model = None

df = pd.DataFrame(results)
df = df.sort_values(by=["Num. trainable params"], ascending=False)
df.index = range(len(df))
df
```
